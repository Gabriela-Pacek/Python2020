Benzylsuccinate synthase (BSS) gene expression level
Data taken from Meyer-Cifuentes et al., 2020 doi: 10.3390/microorganisms8050681

```python
#Attempt #1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Dataset for benzylsuccinate synthase gene expression 

bar_names=["AN_T2_5","AN_T5","AN_T10","AB_TN","AB_T"]
bar_pix=[331.2417582417582, 297.1271585557299, 59.42543171114596,
16.507064364207167, 177.17582417582412]
err_bar_pix=[126.55416012558871, 154.06593406593407, 57.22448979591837, 0,  14.306122448979636]

#Scale correction

bss_fold_Y=["Y_lenght_in_pixels", 552.4364207221349]
scale_factor=bss_fold_Y[-1]/250
print(scale_factor)
bar_corr=[x/scale_factor for x in bar_pix]
print(bar_corr)
anaerobic=["AN_T2_5","AN_T5","AN_T10"]
aerobic=["AB_TN", "AB_T"]
err_corr=[x/scale_factor for x in err_bar_pix]
print(err_corr)
err_lower=[0,0,0,0,0]
Conditions=["anaerobic", "aerobic"]

bss_title='BSS gene expression level in anaerobic and aerobic conditions'
#barchart


plt.figure(figsize=(15, 10))
plt.bar(x=bar_names, height=bar_corr, width=0.7, color=['blue', 'blue', 'blue', 'pink', 'pink'], alpha=0.7, yerr=(err_lower, err_corr))
plt.xlabel("Conditions", fontsize=12, fontweight='bold', color='navy')
plt.ylabel("Fold change - toluene/benzoate grown cells", fontsize=12, fontweight='bold', color='navy')
plt.legend(Conditions)
plt.title(bss_title, fontsize=15, fontweight='bold', color='navy')
plt.show()
```
