```
import matplotlib.pyplot as plt
import matplotlib.axes as max
import numpy as np
import scipy.interpolate
import plotly.graph_objects as go
from scipy import signal
from scipy.interpolate import make_interp_spline, BSpline
from scipy.interpolate import interp1d
```

```
#DATASET
temperature=[20, 30, 40, 50, 60]
activity=[55.143343485518, 93.4199767585044, 122.486210611586, 150.438596390532, 63.5892625043672]
pH=[6.0, 6.5, 7.0, 8.0, 9.0]
act=[0, 52.5490237505116, 28.5457486231482, 93.4199767585044, 73.4188452629058]
y4=[52.33313530866, 41.0788655707634, 61.6646900433926, 42.8078678653819, 38.6931959925441]
x4=[0, 0.5, 1, 2, 4]
y5=[91.241057899921, 80.3860968333016, 72.2224578603028, 45.2048836015532, 26.0947086539673]
x5=[0, 0.5, 1, 2, 4]
y6=[93.2179301067217, 34.7899610039875,12.442311689081]
x6=[0,0.5,1]
y1=[100, 78.4949446053269,117.831063779564,81.7987831474301,73.9363230663179]
x1=[0, 0.5, 1, 2, 4]
y2=[100,88.1029864005678,79.1556559323556,49.5444536067707,28.5997436401819]
x2=[0, 0.5, 1, 2, 4]
y3=[100, 37.3211043885632, 13.3475519943817]
x3=[0,0.5,1]
```

```
#PLOTS - 1st version
plt.figure(figsize=(15,12))
ax0=plt.subplot(2,2,1)
ax0.plot(temperature, activity, color='navy', marker='^')
max.Axes.set_xlabel(ax0, xlabel='Temperature [°C]')
max.Axes.set_ylabel(ax0, ylabel='S.A. [mU/mg]')
max.Axes.set_ylim(ax0, 0)
max.Axes.set_title(ax0, label='Temperature effect on Acmb2 specific activity')

ax1=plt.subplot(2,2,2)
ax1.plot(pH, act, color='navy', marker='^')
max.Axes.set_xlabel(ax1, xlabel='pH')
max.Axes.set_ylabel(ax1, ylabel='S.A. [mU/mg]')
max.Axes.set_ylim(ax1, 0)
max.Axes.set_title(ax1, label='pH effect on Acmb2 specific activity')

ax2=plt.subplot(2,2,3)
plt.plot(x4,y4, color='navy', marker='^')
plt.plot(x5,y5,color='maroon', marker='^')
plt.plot(x6,y6, color='green', marker='^')
ax2.legend(['30°C', '40°C', '50°C'])
max.Axes.set_xlabel(ax2,xlabel="Time [h]")
max.Axes.set_ylabel(ax2,ylabel="S.A. [mU/mg]")
max.Axes.set_title(ax2, label="AcmB2 stability")

ax3=plt.subplot(2,2,4)
plt.plot(x1,y1, color='navy', marker='^')
plt.plot(x2,y2,color='maroon', marker='^')
plt.plot(x3,y3, color='green', marker='^')
ax3.legend(['30°C', '40°C', '50°C'])
max.Axes.set_xlabel(ax3,xlabel="Time [h]")
max.Axes.set_ylabel(ax3,ylabel="Activity [%]")
max.Axes.set_title(ax3, label="AcmB2 activity decrease over time")
```

```
#Temperature plot could really look a little bit smoother...
temp=np.array([20, 30, 40, 50, 60])
xnew1 = np.linspace(temp.min(), temp.max(), 160) 

#define spline
spl = make_interp_spline(temp, activity, k=3)
y_smooth1 = spl(xnew1)

#create smooth line chart 
plt.plot(xnew1, y_smooth1)
plt.show()
```

```
#same with pH plot
pH_np=np.array([6.0, 6.5, 7.0, 8.0, 9.0])
xnew2 = np.linspace(pH_np.min(), pH_np.max(), 150) 

#define spline
spl2 = make_interp_spline(pH_np, act, k=3)
y_smooth2 = spl2(xnew2)

#create smooth line chart 
plt.plot(xnew2, y_smooth2)
plt.show()
```
